```python
import matplotlib.pyplot as plt
import numpy as np
import pyamtrack.stopping as stp
```

```python
energies = np.logspace(np.log10(0.01), np.log10(500), 500)

material = 1
particle_proton = 1001
particle_carbon = 6012
```

```python
results_proton = stp.stopping_power(energies, particle=particle_proton, material=material)
results_carbon = stp.stopping_power(energies, particle=particle_carbon, material=material)
```

```python
plt.plot(energies, results_proton, color="blue", label="Proton")
plt.plot(energies, results_carbon, color="orange", label="Carbon")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Energy (MeV/nucleon)")
plt.ylabel("Stopping Power (MeV*cm²/g)")
plt.legend()
plt.show()
```
