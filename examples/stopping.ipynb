```python
import matplotlib.pyplot as plt
import numpy as np
import pyamtrack.stopping as stp
```

```python
energies = np.logspace(np.log10(0.01), np.log10(500), 500)

material = 1
particle = 1001
```

```python
results = stp.stopping_power(energies, particle=particle, material=material)
```

```python
plt.plot(energies, results)
plt.xscale("log")
plt.yscale("log")
```
